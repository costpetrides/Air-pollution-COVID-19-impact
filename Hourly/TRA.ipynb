```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Hourly/HTRA.csv'
df = pd.read_csv(url)
print(df)
```

```python
fig,ax = plt.subplots(figsize=(5, 3))


NO=[]
x=[]
for i in df['NO20']:
  if i==0:
    x+=[None]
  else:
    x+=[i]
    NO+=[i]

NO2=[]
xx=[]
for i in df['NO220']:
  if i==0:
    xx+=[None]
  else:
    xx+=[i]
    NO2+=[i]

O3=[]
y=[]
for i in df['O320']:
  if i==0:
    y+=[None]
  else:
    y+=[i]
    O3+=[i]



ax.plot(df['HOUR'], x, 'green')
ax.scatter(df['HOUR'], x,color='green', label='$NO$ 2020')
plt.xticks(rotation=90)

ax.plot(df['HOUR'], xx, 'orange')
ax.scatter(df['HOUR'], xx,color='orange', label='$NO_{2}$ 2020')
plt.xticks(rotation=90)


ax2=ax.twinx()
ax2.plot(df['HOUR'], y, 'red')
ax2.scatter(df['HOUR'], y,color='red', label='$O_{3}$ 2020')

plt.xlim(df.loc[0]['HOUR'] , df.loc[23]['HOUR'])
plt.ylim(0,120)
plt.xticks(rotation=90)

ax.set_xlabel('Hours')
ax.set_ylabel(r"$NO_{x}$  $[μg/m^{3}$]")
ax2.set_ylabel(r"$O_{3} \: [μg/m^{3}$]")
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=0)

plt.show()
```
