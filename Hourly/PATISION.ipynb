```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Hourly/HPAT.csv'
df = pd.read_csv(url)
print(df)
```

```python
plt.figure(figsize=(6,3))

plt.xlabel('Hour')
plt.ylabel('$NO_{x}$  $[Î¼g/m^{3}]$')

NO=[]
x=[]
for i in df['NO2O']:
  if i==0:
    x+=[None]
  else:
    x+=[i]
    NO+=[i]

NO2=[]
y=[]
for i in df['NO22O']:
  if i==0:
    y+=[None]
  else:
    y+=[i]
    NO2+=[i]


O3MEAN=[]
xx=[]
for i in df['O320']:
  if i==0:
    xx+=[None]
  else:
    xx+=[i]
    O3MEAN+=[i]

plt.plot(df['HOUR'], xx, 'red')
plt.scatter(df['HOUR'], xx, color='red' ,label='$O_{3}$ Lockdown 2020')
plt.legend()

plt.plot(df['HOUR'], x, 'green')
plt.scatter(df['HOUR'], x, color='green', label='$NO$ Lockdown 2020')

plt.plot(df['HOUR'], y, 'orange')
plt.scatter(df['HOUR'], y,color='orange', label='$NO_{2}$ Lockdown 2020')

plt.legend()

plt.xlim(df.loc[0]['HOUR'] , df.loc[23]['HOUR'])
plt.ylim(0,100)

plt.xticks(rotation=90)

plt.show()
```
