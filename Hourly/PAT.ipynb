```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Hourly/HPAT.csv'
df = pd.read_csv(url)
print(df)
```

```python
plt.figure(figsize=(6,3))

plt.xlabel('Hour')
plt.ylabel('$NO_{x}$  $[Î¼g/m^{3}]$')

NO=[]
x=[]
for i in df['NO19']:
  if i==0:
    x+=[None]
  else:
    x+=[i]
    NO+=[i]

NO2=[]
xx=[]
for i in df['NO219']:
  if i==0:
    xx+=[None]
  else:
    xx+=[i]
    NO2+=[i]

O3=[]
y=[]
for i in df['O319']:
  if i==0:
    y+=[None]
  else:
    y+=[i]
    O3+=[i]

plt.plot(df['Hour'], x, 'green')
plt.scatter(df['Hour'], x, color='green', label='$NO$ 2019')

plt.plot(df['Hour'], xx, 'orange')
plt.scatter(df['Hour'], xx,color='orange', label='$NO_{2}$ 2019')

plt.plot(df['Hour'], y, 'red')
plt.scatter(df['Hour'], y,color='red', label='$O_{3}$ 2019')

plt.legend()

plt.xlim(df.loc[0]['Hour'] , df.loc[23]['Hour'])
plt.ylim(0,120)

plt.xticks(rotation=90)

plt.show()
```

```python
fig,ax = plt.subplots(figsize=(6, 3))


CO=[]
x=[]
for i in df['CO2O']:
  if i==0:
    x+=[None]
  else:
    x+=[i]
    CO+=[i]


O3=[]
y=[]
for i in df['O32O']:
  if i==0:
    y+=[None]
  else:
    y+=[i]
    O3+=[i]



ax.plot(df['Hour'], x, 'purple')
ax.scatter(df['Hour'], x,color='purple', label='$CO$ 2020')
plt.xticks(rotation=90)

ax2=ax.twinx()
ax2.plot(df['Hour'], y, 'red')
ax2.scatter(df['Hour'], y,color='red', label='$O_{3}$ 2020')

plt.xlim(df.loc[0]['Hour'] , df.loc[23]['Hour'])
plt.ylim(0,100)
plt.xticks(rotation=90)

ax.set_xlabel('Hours')
ax.set_ylabel(r"CO  $[mg/m^{3}$]")
ax2.set_ylabel(r"$O_{3} \: [Î¼g/m^{3}$]")
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=0)

plt.show()
```
