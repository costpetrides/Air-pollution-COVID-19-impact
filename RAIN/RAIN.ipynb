```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/RAIN/RAIN.csv'
df = pd.read_csv(url)
print(df)

plt.figure(figsize=(13,2))

plt.xlabel('Days')
plt.ylabel('')



plt.plot(df['DAY'], df['RAIN19'], 'g', label='2020')
plt.plot(df['DAY'], df['RAIN20'], 'b', label='2016-2019')
plt.legend()


plt.xlim(df.loc[0]['DAY'] , df.loc[91]['DAY'])
plt.ylim(0,30)

plt.axvline(x='4-May' ,color='r')
plt.axvline(x='22-Mar' ,color='r')
plt.xticks(rotation=90)

plt.show()
```
