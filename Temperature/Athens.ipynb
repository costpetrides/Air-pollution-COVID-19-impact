```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Temperature/METEO.csv'
df = pd.read_csv(url)
print(df)
```

```python
plt.figure(figsize=(13,2))

plt.xlabel('Days')
plt.ylabel('T [$^{o}C$]')



plt.plot(df['DAY'], df['T2O'], 'g', label='2020')
plt.fill_between(df['DAY'], df['TMIN'], df['TMAX'], alpha=0.2, label='min-max(2016-2019, 2021-2023)')
plt.legend()


plt.xlim(df.loc[0]['DAY'] , df.loc[91]['DAY'])
plt.ylim(0,40)

plt.axvline(x='4-May' ,color='r')
plt.axvline(x='22-Mar' ,color='r')
plt.xticks(rotation=90)

plt.gca().yaxis.set_minor_locator(plt.MultipleLocator(2))
plt.grid(linestyle='--', alpha=0.5)
plt.grid(which='minor', axis='y', linestyle='--', linewidth=0.5)

plt.show()
```
