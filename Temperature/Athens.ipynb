```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Temperature/ATHTEMP.csv'
df = pd.read_csv(url)
print(df)
```

```python
plt.figure(figsize=(13,2))

plt.xlabel('Days')
plt.ylabel('T [$^{o}C$]')



plt.plot(df['DAY'], df['20MEAN'], 'g', label='2020')
plt.plot(df['DAY'], df['19MEAN'], 'b', label='2019')
plt.fill_between(df['DAY'], df['19MIN'], df['19MAX'], alpha=0.2, label='2019 min-max ')
plt.legend()


plt.xlim(df.loc[0]['DAY'] , df.loc[91]['DAY'])
plt.ylim(0,40)

plt.axvline(x='4-May' ,color='r')
plt.axvline(x='22-Mar' ,color='r')
plt.xticks(rotation=90)

plt.show()
```
