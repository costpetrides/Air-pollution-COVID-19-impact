```python
import pandas as pd
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Yearly/Data-Y/Geoponiki.csv'
df = pd.read_csv(url)
print(df)
```

___
**CO**

```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd


array = [[ 0.8, 0.6, 0.4, 0.4, 0.3, 0.4, 0.7, 0.7, 0.8, 0.8, 0.6, 0.9],    # G: years and S: months
         [ 1.2, 0.9, 0.7, 0.7, 0.4, 0.3, 0.3, 0.4, 0.5, 0.6, 0.7, 0.9],
         [ 0.8, 0.7, 0.6, 0.5, 0.5, 0.5, 0.3, 0.3, 0.3, 0.3,  None, 0.3],
         [ 0.5, 0.6, 0.3, 0.3, 0.3, 0.4, 0.5, 0.5,  None, 0.3, 0.6, 0.6]]

month=['Jan', 'Feb', 'Mar*', 'Apr**', 'May*', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov*', 'Dec*']     
year=['2018','2019','2020*','2021']

df_cm = pd.DataFrame(array, index = year, 
                     columns = month)

plt.figure(figsize = (10,7))


plt.title('$CO$')
sn.heatmap(df_cm, annot=True, linewidth=0.01, cbar_kws={'label': 'mg/$m^{3}$'}, cmap='Blues',fmt='g')

plt.savefig('heatmap.png')
plt.savefig('heatmap.pdf')
```

```python
plt.figure(figsize=(17,5))

plt.xlabel('Day')
plt.ylabel('$CO$ mg/$m^{3}$]')
plt.title("$CO$  - Γεωπονική")


plt.bar(df['DAY'], df['CO'], label= 'CO 2018-2021',ec='darkcyan')
plt.legend()


plt.xlim(df.loc[1]['DAY'] , df.loc[1460]['DAY'])
plt.ylim(0,3)

plt.xticks(df['DAY'][::102])

plt.axvline(x='22-Mar-18' ,color='b')
plt.axvline(x='4-May-18' ,color='b')

plt.axvline(x='22-Mar-19' ,color='b')
plt.axvline(x='4-May-19' ,color='b')

plt.axvline(x='22-Mar-20' ,color='r')
plt.axvline(x='4-May-20' ,color='r')

plt.axvline(x='22-Mar-21' ,color='b')
plt.axvline(x='4-May-21' ,color='b')


plt.show()
```
