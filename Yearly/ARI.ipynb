**$SO_{2}$**


```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd


array = [[ 5, 4, 4, 8, 4, 6, 4, 3, 3, 3, 2, 5],
         [ 4, 6, 7, 7, 7, 6, 7, 3, 5, 3, 3, 3],
         [ 4, 3, 4, 7, 5, 6, 5, 3, None, None, None, None ],    # G: years and S: months
         [ None, None, None, None, None, None,  4, 2, 2, 3, 3, 3],
         [ 2, 4, None, 4, 7, 7, 6, 7, 9, None, 3, 6],
         [ 4, 3, 3, 3, 5, 4, 2, 3, 3, 2, 3, 6],
         [ 8, 2, 2, 3, 3, 3, 2, 3, 4, 5, 3, 3],
         [ 3, 4, 3, 2, 2, 4, 6, 6, 6, 3, 5, 6]]

month=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
year=['2016','2017','2018','2019','2020','2021','2022', '2023']

df_cm = pd.DataFrame(array, index = year,
                     columns = month)

plt.figure(figsize = (5,3))

sn.heatmap(df_cm, annot=True, linewidth=0.01, cbar_kws={'label': '$SO_{2} \:\: [μg/m^{3}]$'}, cmap='Blues',fmt='g')

# Highlighting the cell 'April 2020'
target_year = '2020'
target_month = 'Apr'

# Find the index positions of the target year and month
target_year_index = df_cm.index.get_loc(target_year)
target_month_index = df_cm.columns.get_loc(target_month)

# Add a red border to the target cell
plt.gca().add_patch(patches.Rectangle((target_month_index, target_year_index), 1, 1, fill=False, edgecolor='red', lw=1))

plt.savefig('heatmap.png')
plt.savefig('heatmap.pdf')
```
