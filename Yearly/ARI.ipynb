# ARISTOTELOUS U-T (ARI)

*   $PM_{10}$
*   $PM_{2.5}$



**$PM_{10}$**

```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd


array = [[ 38, 31, 47, 42, 36, 32, 33, 33, 31, 40, 32, 40],    # G: years and S: months
         [ 42, 35, 37, 39, 31, 34, 29, 30, 35, 42, 42, 42],
         [ 46, 35, 32, 25, 34, 25, 29, 29, 31, 31, -1, 30], 
         [ 37, 42, 27, 31, 30, 29, 31, 36, 32, 26, 40, 34]]

month=['Jan', 'Feb', 'Mar*', 'Apr**', 'May*', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov*', 'Dec*']      
year=['2018','2019','2020*','2021']

df_cm = pd.DataFrame(array, index = year, 
                     columns = month)

plt.figure(figsize = (10,7))


plt.title('$ΑΣ_{10}$')
sn.heatmap(df_cm, annot=True, linewidth=0.01, cbar_kws={'label': 'μg/$m^{3}$'}, cmap='Blues',fmt='g')

plt.savefig('heatmap.png')
plt.savefig('heatmap.pdf')
```
