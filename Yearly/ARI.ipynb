# ARISTOTELOUS U-T (ARI)




*   $SO_{2}$
*   NO
*   $NO_{2}$
*   $PM_{2.5}$
*   $PM_{10}$




```python
import pandas as pd
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Yearly/Data-Y/Aristotelous.csv'
df = pd.read_csv(url)
print(df)
```

---
**$SO_{2}$**


```python
plt.figure(figsize=(17,5))

plt.xlabel('Day')
plt.ylabel('$SO_{2}$ μg/$m^{3}$]')
plt.title("$SO_{2}$  - Aριστοτέλους")


plt.bar(df['DAY'], df['SO2'], label='$SO_{2}$ 2018-2021',ec='darkcyan')
plt.legend()


plt.xlim(df.loc[1]['DAY'] , df.loc[1460]['DAY'])
plt.ylim(0,20)

plt.xticks(df['DAY'][::102])

plt.axvline(x='22-Mar-18' ,color='b')
plt.axvline(x='4-May-18' ,color='b')

plt.axvline(x='22-Mar-19' ,color='b')
plt.axvline(x='4-May-19' ,color='b')

plt.axvline(x='22-Mar-20' ,color='r')
plt.axvline(x='4-May-20' ,color='r')

plt.axvline(x='22-Mar-21' ,color='b')
plt.axvline(x='4-May-21' ,color='b')

plt.show()
```

**$PM_{2.5}$**

```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd


array = [[ 30,  8, 18, 21, 17, 15, 17, 14, 14, 21, 18, 27],    # G: years and S: months
         [ 24, 24, 20, 21, 16, 17, 15, 14, 16, 23, 20, 24],
         [ 28, 20, 17, 14, 15, 11, 14, 14, 13, 14, 17, 17], 
         [ 20, 25, 16, 17, 15, 17, 16, 22, 16, 14, 24, 20]]

month=['Jan', 'Feb', 'Mar*', 'Apr**', 'May*', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov*', 'Dec*']      
year=['2018','2019','2020*','2021']

df_cm = pd.DataFrame(array, index = year, 
                     columns = month)

plt.figure(figsize = (10,7))


plt.title('$ΑΣ_{2.5}$')
sn.heatmap(df_cm, annot=True, linewidth=0.01, cbar_kws={'label': 'μg/$m^{3}$'}, cmap='Blues',fmt='g')

plt.savefig('heatmap.png')
plt.savefig('heatmap.pdf')
```

```python
plt.figure(figsize=(17,5))

plt.xlabel('Day')
plt.ylabel('$PM_{2.5}$ μg/$m^{3}$]')
plt.title("$PM_{2.5}$  - Αλίαρτος")


plt.bar(df['DAY'], df['PM25'], label='$ΝO_{2}$ 2018-2021', ec='darkcyan')
plt.legend()


plt.xlim(df.loc[1]['DAY'] , df.loc[1460]['DAY'])
plt.ylim(0,85)

plt.xticks(df['DAY'][::102])

plt.axvline(x='22-Mar-18' ,color='b')
plt.axvline(x='4-May-18' ,color='b')

plt.axvline(x='22-Mar-19' ,color='b')
plt.axvline(x='4-May-19' ,color='b')

plt.axvline(x='22-Mar-20' ,color='r')
plt.axvline(x='4-May-20' ,color='r')

plt.axvline(x='22-Mar-21' ,color='b')
plt.axvline(x='4-May-21' ,color='b')


plt.show()
```
