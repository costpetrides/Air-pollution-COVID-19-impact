**$SO_{2}$**

```python
import pandas as pd
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Yearly/Data-Y/Aliartos.csv'
df = pd.read_csv(url)

plt.figure(figsize=(17,5))

plt.xlabel('Day')
plt.ylabel('$SO_{2}$ μg/$m^{3}$]')
plt.title("$SO_{2}$  - Πατησίων")


plt.bar(df['DAY'], df['SO2'], label='$SO_{2}$ 2018-2021',ec='darkcyan')
plt.legend()


plt.xlim(df.loc[1]['DAY'] , df.loc[1460]['DAY'])
plt.ylim(0,50)

plt.xticks(df['DAY'][::102])

plt.axvline(x='22-Mar-18' ,color='b')
plt.axvline(x='4-May-18' ,color='b')

plt.axvline(x='22-Mar-19' ,color='b')
plt.axvline(x='4-May-19' ,color='b')

plt.axvline(x='22-Mar-20' ,color='r')
plt.axvline(x='4-May-20' ,color='r')

plt.axvline(x='22-Mar-21' ,color='b')
plt.axvline(x='4-May-21' ,color='b')


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Yearly/Data-Y/Aliartos.csv'
df = pd.read_csv(url)

plt.figure(figsize=(17,5))

plt.xlabel('Day')
plt.ylabel('$SO_{2}$ μg/$m^{3}$]')
plt.title("$SO_{2}$  - Πατησίων")


x=[]
for i in  df['SO2']:
  if i==0:
    x+=[None]
  else:
    x+=[i]



plt.plot(df['DAY'], x, label='$SO_{2}$ 2018-2021')
plt.legend()


plt.xlim(df.loc[1]['DAY'] , df.loc[1460]['DAY'])
plt.ylim(0,50)

plt.xticks(df['DAY'][::102])

plt.axvline(x='22-Mar-18' ,color='g')
plt.axvline(x='4-May-18' ,color='g')

plt.axvline(x='22-Mar-19' ,color='g')
plt.axvline(x='4-May-19' ,color='g')

plt.axvline(x='22-Mar-20' ,color='r')
plt.axvline(x='4-May-20' ,color='r')

plt.axvline(x='22-Mar-21' ,color='g')
plt.axvline(x='4-May-21' ,color='g')


plt.show()
```
