**$Ο_{3}$**

```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd

array = [[ 76, 76, 79, 87, 93, 96, 109, 117, 94, 74, 63, 62],
         [ 63, 72, 84, 87, 95, 93, 119, 126, 93, 78, 59, 63],
         [ 59, 66, 79, 106, 96, 101, 101, 103, 96, 73, 58, 52],    # G: years and S: months
         [ 48, None, None, None, 75, 88, 104, 109, 98, 74, 63, 55],
         [ 64, 70, 72, 86, 84, 81, 96, 97, 87, 67, 61, 50],
         [ 50, 57, 66, 66, 88, 105, 115, 113, 96, 71, 58, 57]]
        

month=['Jan', 'Feb', 'Mar*', 'Apr**', 'May*', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov*', 'Dec*']      
year=['2017','2016','2018','2019','2020*','2021']

df_cm = pd.DataFrame(array, index = year, 
                     columns = month)

plt.figure(figsize = (10,7))


plt.title('$Ο_{3}$')
sn.heatmap(df_cm, annot=True, linewidth=0.01, cbar_kws={'label': 'μg/$m^{3}$'}, cmap='Blues',fmt='g')

plt.savefig('heatmap.png')
plt.savefig('heatmap.pdf')
```
