```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd


array = [[ 65, 71, 85, 100, 105, 105, 112, 121, 103, 80, 71, 71],
         [ 74, 39, 100, 97, 102, 105, 113, 124, 100, 87, 77, 68],
         [ 71, 91, 101, 115, 111, 110, 110, 111, 101, 76, 70, 54],    # G: years and S: months
         [ 63, 70, 72, None, 98, 116, 118, 119, 107, 89, 72, 59],
         [ 67, 64, 81,  95, 97, 86, 98, 105, 92, 77, None, 61],
         [ 68, 74, 83, 90, 95, 105, 115, 118, 98, 76, 65, 62]]

month=['Jan', 'Feb', 'Mar*', 'Apr**', 'May*', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov*', 'Dec*']
year=['2017','2016','2018','2019','2020*','2021']

df_cm = pd.DataFrame(array, index = year,
                     columns = month)

plt.figure(figsize = (5,3))


sn.heatmap(df_cm, annot=True, linewidth=0.01, cbar_kws={'label':  '$O_{3} \:\: Î¼g/m^{3}$'}, cmap='Blues',fmt='g')

plt.savefig('heatmap.png')
plt.savefig('heatmap.pdf')
```
