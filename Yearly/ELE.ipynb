# Elefsinas I-S (ELE)

*   $PM_{10}$
*   $PM_{2.5}$


**$PM_{10}$**

```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd


array = [[ 27, 23, 41, 39, 30, 29, 28, 26, 25, 31, 23, 25],    # G: years and S: months
         [ 26, 24, 25, 27, 22, 25, 23, 25, 26, 28, 26, 24],
         [ 26, 23, 22, 21, 27, 19, 23, 25, 23, 20, 17, 23], 
         [ 28, 30, 20, 25, 25, 28, 27, 33, 28, 22, 32, 24]]

month=['Jan', 'Feb', 'Mar*', 'Apr**', 'May*', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov*', 'Dec*']      
year=['2018','2019','2020*','2021']

df_cm = pd.DataFrame(array, index = year, 
                     columns = month)

plt.figure(figsize = (10,7))


plt.title('$ΑΣ_{10}$')
sn.heatmap(df_cm, annot=True, linewidth=0.01, cbar_kws={'label': 'μg/$m^{3}$'}, cmap='Blues',fmt='g')

plt.savefig('heatmap.png')
plt.savefig('heatmap.pdf')
```
