```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd


array = [[ 9, 9, 8, 12, 11, 12, 9, 5, 8, 5, None, None],    # G: years and S: months
         [ None, None, 5, 6, 8, 10, 8, 6, 6, 11, 10, 10],
         [ 16, 18, 14, 4, 4, 6, 4, 4, 5, 6, 6, 8],
         [ 7, 4, 4, 4, 6, 9, 5, 5, 6, 5, 7, 9]]

month=['Jan', 'Feb', 'Mar*', 'Apr**', 'May*', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov*', 'Dec*']      
year=['2018','2019','2020*','2021']

df_cm = pd.DataFrame(array, index = year, 
                     columns = month)

plt.figure(figsize = (10,7))


plt.title('$SO_{2}$')
sn.heatmap(df_cm, annot=True, linewidth=0.01, cbar_kws={'label': 'Î¼g/$m^{3}$'}, cmap='Blues',fmt='g')

plt.savefig('heatmap.png')
plt.savefig('heatmap.pdf')
```
