**$O_{3}$**

```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd


array = [[ 41, 43, 67, 74, 87, 90,  97,  96, 75, 56, 37, 48],
         [ 44, 44, 65, 72, 82, 80,  95, 105, 83, 59, 39, 37],
         [ 35, 63, 70, 81, 80, 83,  86,  88, 79, 53, 37, 30],    # G: years and S: months
         [ 43, 51, 56, 56, 59, 90,  97, 105, 85, 56, 40, 16],
         [ 40, 55, 63, 81, 78, 77,  96,  93,  78, 50, 45, 39],
         [ 43, 53, 72, 80, 90, 94, 106, 103, 85, 59, 49, 45]]

month=['Jan', 'Feb', 'Mar*', 'Apr**', 'May*', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov*', 'Dec*']      
year=['2016','2017','2018','2019','2020*','2021']

df_cm = pd.DataFrame(array, index = year, 
                     columns = month)

plt.figure(figsize = (10,7))


plt.title('$Ο_{3}$')
sn.heatmap(df_cm, annot=True, linewidth=0.01, cbar_kws={'label': 'μg/$m^{3}$'}, cmap='Blues',fmt='g')

plt.savefig('heatmap.png')
plt.savefig('heatmap.pdf')
```
