```python
import scipy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
import matplotlib.ticker as ticker
from scipy.stats import pearsonr
from matplotlib.ticker import ScalarFormatter

url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Lockdown%201/DATA/Thrakomakedones.csv'
df = pd.read_csv(url)
print(df)
```

$NO_{2}$

```python
import matplotlib.pyplot as plt

# Create a figure with a single x-axis
fig, ax1 = plt.subplots(figsize=(13, 2))

# Label the primary y-axis
ax1.set_xlabel('Days')
ax1.set_ylabel('$NO_{2}$ $[μg/m^{3}]$')

# Plot the first dataset on the primary y-axis
y1 = []
for i in df['NO2']:
    if i == 0:
        y1.append(None)
    else:
        y1.append(i)
ax1.plot(df['Day'], y1, 'b', label='2020 $NO$')

# Create a secondary y-axis
ax2 = ax1.twinx()

# Label the secondary y-axis
ax2.set_ylabel('$O_{3}$ $[μg/m^{3}]$')

# Plot the second dataset on the secondary y-axis
ax2.plot(df['Day'], df['O3'], 'g', label='2020 $O_{3}$')

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
ax1.legend(lines, labels, loc='upper left')

# Set the x-axis limits and other formatting
ax1.set_xlim(df.loc[0]['Day'], df.loc[91]['Day'])
ax1.set_ylim(0, 50)
ax1.axvline(x='4-May', color='r')
ax1.axvline(x='22-Mar', color='r')
ax1.tick_params(axis='x', rotation=90)

# Show the plot
plt.show()
```

```python
r=scipy.stats.pearsonr(df['O3'][0:23],df['NO2'][0:23])[0]

plt.figure(figsize=(5,3))
plt.xlabel('$O_{3}$ $[μg/m^{3}]$')
plt.ylabel('$NO_{2}$ $[μg/m^{3}]$')

x=[]
for i in df['O3'][0:23]:
  if i==0:
    x+=[None]
  else:
    x+=[i]

y=[]
for i in df['NO2'][0:23]:
  if i==0:
    y+=[None]
  else:
    y+=[i]

plt.scatter(x, y, label='2020')
plt.legend()

m, b = np.polyfit(df['O3'][0:23],df['NO2'][0:23], 1)

plt.xticks(np.arange(60,125,5))
plt.yticks(np.arange(0,60,5))

plt.xlim(60,120)
plt.ylim(0,25)

xx=np.linspace(10,200,40)
plt.plot(xx, m*xx + b, label='r=%.3f' %(r))
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import scipy.stats

# Assuming you have a DataFrame 'df' with columns 'O3' and 'NO'

# Filter out zeros and None values from 'O3' and 'NO' columns
filtered_df = df[(df['O3'] != 0) & (df['NO'] != 0)].head(23)

# Extract the filtered data into x and y arrays
x = filtered_df['O3'].to_numpy()
y = filtered_df['NO'].to_numpy()

# Calculate Pearson correlation coefficient (r)
r = scipy.stats.pearsonr(x, y)[0]

# Create a new figure for your plot
plt.figure(figsize=(5, 3))

# Set labels for x and y axes
plt.xlabel('$O_{3}$ $[μg/m^{3}]$')
plt.ylabel('$NO$ $[μg/m^{3}]$')

# Create a scatter plot of the filtered data
plt.scatter(x, y, label='2020')

# Fit a linear regression line
m, b = np.polyfit(x, y, 1)

# Set ticks and limits for the x-axis
plt.xticks(np.arange(60, 125, 5))
plt.xlim(60, 120)

# Generate x values for the linear regression line
xx = np.linspace(10, 200, 40)

# Plot the linear regression line and add the correlation coefficient to the legend
plt.plot(xx, m * xx + b, label=f'r={r:.3f}')

# Create a custom y-axis formatter to multiply values by 10^(-1)
def custom_y_format(value, tick_number):
    if value == 0:
        return "0"
    else:
        return str(int(value * 10))

# Apply the custom y-axis formatter
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(custom_y_format))

# Add "10^(-1)" label at the top of the y-axis
plt.text(0.03, 1.05, '$10^{-1}$', transform=plt.gca().transAxes, fontsize=12, va='bottom')

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()
```
