# Athinas

```python
import scipy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Lockdown%201/DATA/AthinasLock.csv'
df = pd.read_csv(url)
print(df)
```

# $NO_{2}$

```python
plt.figure(figsize=(13,2))

plt.xlabel('Days')
plt.ylabel('$NO_{2}$ / $O_{3}$   $[Î¼g/m^{3}]$')

plt.plot(df['DAY'], df['NO220'], 'b', label='2020 $NO_{2}$')
plt.plot(df['DAY'], df['O320'], 'g', label='2020 $O_{3}$')

plt.legend()

plt.xlim(df.loc[0]['DAY'] , df.loc[91]['DAY'])
plt.ylim(0,150)

plt.axvline(x='4-May' ,color='r')
plt.axvline(x='22-Mar' ,color='r')
plt.xticks(rotation=90)

plt.show()
```

```python
r=scipy.stats.pearsonr(df['O320'][0:23],df['NO220'][0:23])[0]

plt.figure(figsize=(5,3))
plt.xlabel('$O_{3}$ $[Î¼g/m^{3}]$')
plt.ylabel('$NO_{2}$ $[Î¼g/m^{3}]$')

x=[]
for i in df['O320'][0:23]:
  if i==0:
    x+=[None]
  else:
    x+=[i]

y=[]
for i in df['NO220'][0:23]:
  if i==0:
    y+=[None]
  else:
    y+=[i]

plt.scatter(x, y, label='2020')
plt.legend()

m, b = np.polyfit(df['O320'][0:23],df['NO220'][0:23], 1)
plt.xticks(np.arange(0,65,5))

plt.ylim(0,40)
xx=np.linspace(0,60,40)
plt.plot(xx, m*xx + b, label='r=%.3f' %(r))
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.xlabel('$O_{3}$  $[Î¼g/m^{3}]$')
plt.ylabel('$NO_{2}$ $[Î¼g/m^{3}]$')

x=[]
for i in df['O320'][23:66]:
  if i==0:
    x+=[None]
  else:
    x+=[i]

y=[]
for i in df['NO220'][23:66]:
  if i==0:
    y+=[None]
  else:
    y+=[i]

r=scipy.stats.pearsonr(x,y)[0]

plt.xticks(np.arange(0,75,5))


plt.scatter(x, y, label='2020')
plt.legend()

m, b = np.polyfit(x, y, 1)

xx=np.linspace(20,70,300)
plt.plot(xx, m*xx + b, label='r=%.3f' %(r))
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.xlabel('$O_{3}$ $[Î¼g/m^{3}]$')
plt.ylabel('$NO_{2}$   $[Î¼g/m^{3}]$')

x=[]
for i in df['O320'][66:91]:
  if i==0:
    x+=[None]
  else:
    x+=[i]

y=[]
for i in df['NO220'][66:91]:
  if i==0:
    y+=[None]
  else:
    y+=[i]

r=scipy.stats.pearsonr(x,y)[0]

plt.scatter(x, y, label='2020')
plt.legend()


plt.xticks(np.arange(0,55,5))
r=scipy.stats.pearsonr(x,y)[0]
m, b = np.polyfit(x,y, 1)

xx=np.linspace(2,50,30)
plt.ylim(0,30)
plt.plot(xx, m*xx + b, label='r=%.3f' %(r))
plt.legend()

plt.show()
```
