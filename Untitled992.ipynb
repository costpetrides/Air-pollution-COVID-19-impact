```python
import matplotlib.pyplot as plt

x_points = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012]
y_points = [10, 15, 13, 14, 20, 26, 27, 27, 27, 20, 17, 14, 13]

plt.figure(figsize=(10, 6)) # Added for better clarity
plt.plot(x_points, y_points, marker='o', linestyle='-', color='blue')

plt.title('CO')
plt.xlabel('Year')
plt.ylabel('ppb')
plt.grid(True, linestyle='--', alpha=0.6)

# Adjusted xlim so the data isn't squashed to the far right
plt.xlim(1999, 2013)
plt.ylim(0, 30)

plt.xticks(range(2000, 2013, 1))
plt.yticks(range(0, 32, 2))

plt.legend() # This now has a label to display!

plt.show()
```
