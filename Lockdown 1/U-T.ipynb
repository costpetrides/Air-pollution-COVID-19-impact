------------------------------

**Patision**




```python
import pandas as pd
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Lockdown%201/PSO2.csv'
df = pd.read_csv(url)



plt.figure(figsize=(17,5))
plt.title("$SO_{2}$ - Πατησίων")

plt.xlabel('year')
plt.ylabel('$SO_{2}$ [μg/m^{3}$]')

plt.title("$SO_{2}$ - Πατησίων")

plt.plot(df['DAY'], df['2020'], 'g', label='2020')
plt.plot(df['DAY'], df['2021'], 'c', label='2021')
plt.plot(df['DAY'], df['2018'], 'b', label='2018')
plt.legend()


plt.xlim(df.loc[10]['DAY'] , df.loc[90]['DAY'])
plt.axvline(x='4-May' ,color='r')
plt.axvline(x='22-Mar' ,color='r')
plt.xticks(rotation=90)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Lockdown%201/PNO2.csv'
df = pd.read_csv(url)

plt.figure(figsize=(17,5))

plt.xlabel('year')
plt.ylabel('$NO_{2}$ [μg/$m^{3}$]')

plt.title("$NO_{2}$ - Πατησίων")

plt.plot(df['DAY'], df['2020'], 'g', label='2020')
plt.plot(df['DAY'], df['2021'], 'c', label='2021')
plt.plot(df['DAY'], df['2019'], 'b', label='2019')
plt.legend()

plt.xlim(df.loc[10]['DAY'] , df.loc[90]['DAY'])

plt.axvline(x='4-May' ,color='r')
plt.axvline(x='22-Mar' ,color='r')

plt.xticks(rotation=90)
plt.show()
```
