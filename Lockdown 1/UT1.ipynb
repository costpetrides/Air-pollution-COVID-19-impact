# **First Lockdown / Urban - Traffic**

**(22 March - 4 May)**

# Patision

```python
import pandas as pd
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/costpetrides/Air-pollution-COVID-19-impact/main/Lockdown%201/DATA/PatisionL1.csv'
df = pd.read_csv(url)
print(df)
```



---


**$SO_{2}$**

```python
plt.figure(figsize=(18,7))
plt.title("$SO_{2}$ - Πατησίων")

plt.xlabel('Days')
plt.ylabel('$SO_{2} \: [μg/m^{3}$]')

plt.title("$SO_{2}$ - Πατησίων")

plt.plot(df['DAY'], df['SO221'], 'c', label='2021')
plt.plot(df['DAY'], df['SO220'], 'g', label='2020')
# plt.plot(df['DAY'], df['SO219'], 'b', label='2019')
plt.plot(df['DAY'], df['SO218'], 'y', label='2018')

plt.legend()


plt.xlim(df.loc[0]['DAY'] , df.loc[91]['DAY'])
plt.ylim(0,25)

plt.axvline(x='4-May' ,color='r')
plt.axvline(x='22-Mar' ,color='r')
plt.xticks(rotation=90)

plt.show()
```
